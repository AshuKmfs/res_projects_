# Train Test data

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size = 0.2,random_state=101)
for train_i, test_i in split.split(df,df['Bankrupt?']):
    train_set = df.loc[train_i]
    test_set= df.loc[test_i]
```

```python
X_train = train_set.drop(['Bankrupt?'],axis = 1)
y_train = train_set['Bankrupt?'].values
```

```python
X_test = test_set.drop(['Bankrupt?'],axis = 1)
y_test = test_set['Bankrupt?'].values
```

```python
from sklearn.preprocessing import MinMaxScaler 
sc = MinMaxScaler()
x_train = sc.fit_transform(X_train)
x_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier =  LogisticRegression(max_iter=1000,penalty='l1', class_weight='balanced', solver='liblinear',random_state = 0)
########################################################
#max_iter: sets the maximum number of iterations for the solver to converge. solver will stop iterating after 1000 iterations
#penalty: This parameter specifies the regularization penalty to be applied to the model.
#class_weight: specifies the weight to be given to each class in the classification problem.'balanced' option means,class weights will be automatically adjusted based on the number of samples in each class. useful when the classes are imbalanced
#solver: specifies the optimization algorithm to be used to find the coefficients of the logistic regression model
#######################################################
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
y_prob = classifier.predict_proba(x_test)[:, 1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)
```

```python
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

```python
import os
```

```python
actors = os.listdir('Actors_data')
print(actors)
```

```python
filenames = []
for actor in actors:
    for file in os.listdir(os.path.join('Actors_data',actor)):
        filenames.append(os.path.join('Actors_data',actor,file))
#print(filenames)
print(len(filenames))
```

```python
import pickle
pickle.dump(filenames,open('filenames.pkl','wb'))
```

```python
import keras
import tensorflow as tf
print(keras.__version__)
print(tf.__version__)
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
import numpy as np
```

```python
model = ResNet50(include_top=False,
             input_shape=(224,224,3),pooling='avg')
```

```python
print(model.summary())
```

```python
def feature_extractor(img_path,model):
    img = image.load_img(img_path,target_size=(224,224))
    img_array = image.img_to_array(img)
    expanded_img = np.expand_dims(img_array,axis=0) # from single image to batch of single image
    preprocessed_img = preprocess_input(expanded_img)
    result = model.predict(preprocessed_img).flatten()
    return result
```

```python
from tqdm import tqdm
features = []
for file in tqdm(filenames):
    features.append(feature_extractor(file,model))
```

```python
pickle.dump(features,open('embeddings.pkl','wb'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import cv2
from mtcnn import MTCNN
from PIL import Image
import matplotlib.pyplot as plt 
```

```python
feature_list = np.array(pickle.load(open('embeddings.pkl','rb')))
filenames = pickle.load(open('filenames.pkl','rb'))
```

```python
detector = MTCNN()
```

```python
sample_img = cv2.imread('sample/100-min.jpg')
results = detector.detect_faces(sample_img)
x,y,width,height = results[0]['box']
print(x,y,width,height)
```

```python
face = sample_img[y:y+height,x:x+width]
cv2.imshow('output',face)
cv2.waitKey(0)
```

```python
# extracting feature
image = Image.fromarray(face)
image = image.resize((224,224))
face_array = np.asarray(image)
face_array = face_array.astype('float32')
```

```python
expanded_img = np.expand_dims(face_array,axis=0)
preprocessed_img = keras.applications.efficientnet.preprocess_input(expanded_img)
result = model.predict(preprocessed_img).flatten()
print(result)
print(result.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print(cosine_similarity(result.reshape(1,-1),feature_list[0].reshape(1,-1))[0][0])
```

```python
similarity = []

for i in range(len(feature_list)):
    similarity.append(cosine_similarity(result.reshape(1,-1),feature_list[i].reshape(1,-1))[0][0])
index_positions = sorted(list(enumerate(similarity)),reverse=True,key=lambda x:x[1])[0][0]
print(index_positions)
```

```python
pred = cv2.imread(filenames[index_positions])
cv2.imshow('output',pred)
cv2.waitKey(0)
```
